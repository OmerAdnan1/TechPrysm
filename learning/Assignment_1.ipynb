# Assignment 1 Linear Regression

```python
import numpy as np
import pandas as pd

np.random.seed(42)

num_homes = 100
area = np.random.randint(800, 4000, size=num_homes)
bedrooms = np.random.randint(1, 6, size=num_homes)
age = np.random.randint(1, 40, size=num_homes)
noise = np.random.normal(0, 15000, size=num_homes)
price = (area * 150) + (bedrooms * 25000) - (age * 1200) + 50000 + noise


df = pd.DataFrame({
    'Area_SqFt': area,
    'Bedrooms': bedrooms,
    'Age_Years': age,
    'Price_USD': price
})

print("Raw Data")
print(df.head(), "\n")

# 1. Separate Features (X) and Target Variable (y)
X = df[['Area_SqFt', 'Bedrooms', 'Age_Years']].values
y = df['Price_USD'].values

# 2. Train-Test Split (80% Train, 20% Test)
split_idx = int(num_homes * 0.8)

X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

class ScratchLinearRegression:
    def __init__(self):
        self.theta = None
        self.intercept_ = None
        self.coef_ = None

    def fit(self, X, y):
        # Add a column of ones to X to handle the bias/intercept term (w0)
        ones_column = np.ones((X.shape[0], 1))
        X_b = np.concatenate((ones_column, X), axis=1)

        # (X^T * X)^(-1) * X^T * y
        self.theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

        # Split theta into intercept and weights for readability
        self.intercept_ = self.theta[0]
        self.coef_ = self.theta[1:]

    def predict(self, X):
        ones_column = np.ones((X.shape[0], 1))
        X_b = np.concatenate((ones_column, X), axis=1)
        return X_b @ self.theta

model = ScratchLinearRegression()
model.fit(X_train, y_train)

#  EVALUATION
# =====================================================================
predictions = model.predict(X_test)

# Calculate Mean Squared Error (MSE)
mse = np.mean((y_test - predictions) ** 2)
# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
# Calculate R-squared (R2)
y_mean = np.mean(y_test)
total_sum_of_squares = np.sum((y_test - y_mean) ** 2)
residual_sum_of_squares = np.sum((y_test - predictions) ** 2)
r2 = 1 - (residual_sum_of_squares / total_sum_of_squares)

# RESULTS
print("--- MODEL PARAMETERS ---")
print(f"Intercept (Base Price): ${model.intercept_:,.2f}")
print(f"Coefficients (Weights):")
print(f"  * Area_SqFt Weight:  ${model.coef_[0]:,.2f} per sqft")
print(f"  * Bedrooms Weight:   ${model.coef_[1]:,.2f} per bedroom")
print(f"  * Age_Years Weight:  ${model.coef_[2]:,.2f} per year of age")
print("-" * 35)
print("\n--- MODEL PERFORMANCE ---")
print(f"Mean Squared Error (MSE):       {mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): ${rmse:,.2f} (Average prediction error)")
print(f"R-squared (R2) Score:           {r2:.4f}")
```
