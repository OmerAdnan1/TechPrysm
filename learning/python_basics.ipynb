```python
# TRY BLOCK

try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
else:
    print("Runs if no exception occurred")
finally:
    print("Always runs — cleanup code goes here")


try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    print("That wasn't a valid number")
except ZeroDivisionError:
    print("Can't divide by zero")
except Exception as e:          # catches anything else — use sparingly
    print(f"Unexpected error: {e}")
else:
    print("Everything is fine")
finally:
    print("Always print this ")


class InsufficientFundsError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(f"Cannot withdraw {amount}, balance is {balance}")
    return balance - amount
```

# Lambda Function
#### IT is basically a one liner function that allows you basic functionality no if else


```python
# Example 1
square = lambda x: x ** 2
print(square(7))

# Example 2
students = [("Ali", 85), ("Sara", 92), ("Zain", 78)]
students.sort(key=lambda x: x[1])          # sort by score
# [('Zain', 78), ('Ali', 85), ('Sara', 92)]
print(students)
```

# Partial Functions

this basically helps you define a function that is already defined. freezing some args

Useful when you're repeatedly calling a function with mostly-fixed arguments — common in callbacks, ML training loops (fixing hyperparams), and functional-style pipelines.

```python
from functools import partial

def power(base, exponent):
    return base ** exponent

square = partial(power, exponent=2)
cube = partial(power, exponent=3)

print("square of 5 = ",square(5))   # 25
print("cube of 5 = ",cube(5))   # 25
```

# Gorilla Patching or Monkey Patching
This basically overrides the function of a class.
can be used for both classes and induvidual instances

```python
# external_api.py (Imagine this is a library you can't easily edit)
import time

class APIClient:
    def fetch_data(self):
        # Pretend this is a slow network request
        time.sleep(2)
        return "Real, expensive data from the cloud"

# 1. Define the replacement behavior
def mock_fetch_data(self):
    return "Fake, super-fast test data"

# 2. Apply the Monkey Patch!
# We are physically swapping the method on the class at runtime.
APIClient.fetch_data = mock_fetch_data

# 3. Test the behavior
client = APIClient()
print(client.fetch_data())
# Output: "Fake, super-fast test data" (No 2-second delay!)
```
